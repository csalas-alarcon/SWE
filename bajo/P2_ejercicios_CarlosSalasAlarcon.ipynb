```python
# Ejercicio 1
def rounded(num: float, cif: int) -> float | int:
    if num == 0:
            return 0
    return round(num, cif - int(math.floor(math.log10(abs(num)))) - 1)
```

```python
# Ejercicio 2
def err_absolute(num: float | int, rounded_num: float | int) -> float:
    return abs((num - rounded_num))
```

```python
# Ejercicio 3 
def err_relative(num: float | int, rounded_num: float | int) -> float:
    return err_absolute(num, rounded_num) / num
```

```python
# Ejercicio 5
def err_arithmetic_array_relative(nums_array: list[float | int], approximated_array: list[float | int]) -> float:
    total_nums: float | int = reduce(lambda x, y: x + y, nums_array)
    total_approximated: float | int = reduce(lambda x, y: x + y, approximated_array)
    
    return err_absolute(total_nums, total_approximated) / total_nums
```

```python
# Ejercicio 6
def err_geometric_array_relative(nums_array: list[float | int], approximated_array: list[float | int]) -> float:
    for a, b in zip(nums_array, approximated_array):
        err_relative += err_relative(a, b)
    
    return err_relative
```

```python
# Ejercicio 7
def err_quotient_array_relative(dividend: float | int, divisor: float | int, cif: int) -> float:
    nums: tuple[float | int] = (dividend, divisor)
    approximated_nums: tuple[int | float] = (rounded(dividend, cif), rounded(divisor, cif))
    for a, b in zip(nums, approximated_nums):
        err_relative += err_relative(a, b)
    
    return err_relative
```
